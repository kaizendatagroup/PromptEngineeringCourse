<h2 align="right">گسترش دادن</h2>
<h4 align="right">.در این درس، ایمیل‌های خدمات مشتری را ایجاد می‌کنید که متناسب با نظر هر مشتری است</h4>

````python
import openai
import os
from dotenv import load_dotenv, find_dotenv

load_dotenv(find_dotenv())


openai.api_key  = os.getenv('API_KEY')
````

````python
def get_completion(prompt, model="gpt-3.5-turbo",temperature=0):
    messages = [{"role": "user", "content": prompt}]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=temperature,
    )
    return response.choices[0].message["content"]
````

````python
sentiment = "negative"

review = f"""
آنها هنوز سیستم 17 قطعه را در فروش فصلی با قیمت حدود 49 دلار در ماه نوامبر داشتند که تقریباً نصف شد،
اما به دلایلی (که به آن افزایش قیمت می گویند) حدود هفته دوم دسامبر،
قیمتها تقریباً از دسامبر بین 70 تا 89 دلار برای همان سیستم افزایش یافت  
،و مدل 11 قطعه ای حدود 10 دلار یا بیشتر از قیمت فروش قبلی 29 دلار افزایش یافت
،بنابراین خوب به نظر می رسد، اما اگر به پایه نگاه کنید
،قسمتی که تیغه در جای خود قفل می شود به خوبی در نسخه های قبلی چند سال پیش به نظر نمی رسد
اما من قصد دارم با آن بسیار ملایم باشم(به عنوان مثال
من اقلام بسیار سفت مانند لوبیا، یخ، برنج و غیره را خرد میکنم.
ابتدا در مخلوط کن،
سپس آنها را به اندازه لازم که می خواهم در مخلوط کن پودر کنید،
سپس به تیغه هم زدنی بروید تا آرد ریزتر شود،
،و هنگام درست کردن اسموتی ها ابتدا از تیغه برش متقاطع استفاده کنید
.سپس در صورت نیاز به آنها از تیغه صاف استفاده کنید
"""
````

````python
prompt = f"""
.شما یک دستیار هوش مصنوعی  فارسی زبان خدمات مشتری هستید
.وظیفه شما ارسال یک پاسخ ایمیل به یک مشتری با ارزش است
،با توجه به ایمیل مشتری که با ``` مشخص شده است
.پاسخی برای تشکر از مشتری برای بررسی آنها ایجاد کنید
،اگر این احساس مثبت یا خنثی است، از آنها برای بررسی آنها تشکر کنید
اگر این احساس منفی است، عذرخواهی کنید و پیشنهاد دهید که 
!آنها با خدمات مشتری تماس بگیرند
.اطمینان حاصل کنید که از جزئیات خاصی از بررسی استفاده می کنید
.با لحن مختصر و حرفه ای بنویسید
.ایمیل را به عنوان «نماینده مشتری هوش مصنوعی» امضا کنید
Customer review: ```{review}```
Review sentiment: {sentiment}
"""
response = get_completion(prompt)
print(response)
````

<h2 align="right">مدل را یادآوری کنید تا از جزئیات ایمیل مشتری استفاده کند</h2>

````python
prompt = f"""
شما یک دستیار هوش مصنوعی فارسی زبان خدمات مشتری هستید.
وظیفه شما ارسال یک پاسخ ایمیل به یک مشتری با ارزش است.
با توجه به ایمیل مشتری که با ``` مشخص شده است،
پاسخی ایجاد کنید تا از نظر مشتری تشکر کنید.
اگر این احساس مثبت یا خنثی است، از آنها برای بررسی آنها تشکر کنید.
اگر این احساس منفی است، عذرخواهی کنید و پیشنهاد کنید که آنها بتوانند با خدمات مشتری تماس بگیرند.
اطمینان حاصل کنید که از جزئیات خاصی از بررسی استفاده می کنید. با لحن مختصر و حرفه ای بنویسید.
ایمیل را به عنوان «نماینده مشتری هوش مصنوعی» امضا کنید.
Customer review: ```{review}```
Review sentiment: {sentiment}
"""
response = get_completion(prompt, temperature=0.7)
print(response)
````
